```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# Define BitsAndBytesConfig for 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,  # Enable 4-bit quantization
    bnb_4bit_quant_type="nf4",  # Normalized float-4 (common for better precision)
    bnb_4bit_use_double_quant=True,  # Double quantization for memory efficiency
    bnb_4bit_compute_dtype="float16"  # Use float16 computation
)

# Load model and tokenizer
model_name = "Thomasche/deepseek-leetcode"  # Replace with the correct model name if needed
model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Create text-generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Generate textUnused kwargs: ['_load_in_4bit', '_load_in_8bit', 'quant_method']. These kwargs are not used in <class 'transformers.utils.quantization_config.BitsAndBytesConfig'>.
output = generator("Write a Python function to check if a number is prime:", max_length=2000)
print(output[0]["generated_text"])
```

```python
def isPrime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
```

```python
# Generate a response
output = generator("Generate code for the given problem in Python: Given a string s, find the length of the longest substring without repeating characters.",max_length=4000)
print(output[0]["generated_text"])
```
